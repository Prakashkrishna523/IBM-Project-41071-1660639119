## 9. Generate the series of dates from 1st Jan, 2023 to 10th Feb, 2023

```python
#Generate the series of dates from 1st Jan, 2023 to 10th Feb, 2023
import pandas as pd
per1 = pd.date_range(start ='01-01-2023',
		end ='02-10-2023')

for val in per1:
	print(val)

```
